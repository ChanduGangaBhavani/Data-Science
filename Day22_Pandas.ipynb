## Series Data

- list
- tuple
- numpy array

```python
s1=pd.Series([1,2,3,"apssdc"])
print(s1)
print(type(s1))
```

```python
s1=pd.Series([1,2,3,"apssdc"],index=['x1','x2','x3','x4'])
print(s1)
print(type(s1))
```

```python
s2=pd.Series((1,2,3,4,5)) ##tuple
print(s2)
print(type(s2))
s2.index=[11,12,13,14,15]
print(s2)
```

```python
## Series object with numpy array

import numpy as np
s3=pd.Series(np.arange(1,10))
print(s3)
print(type(s3))
```

## pandas date_range
- to work with date and time formats we can use data_range
- pandas.date_range()

```python
import pandas as pd
d1=pd.date_range('01-01-2021',periods=10)
print(d1)
```

```python
import pandas as pd
d1=pd.date_range('01-01-2021',periods=20,freq='1H')
print(d1)
```

```python
import pandas as pd
d1=pd.date_range('01-01-2021',periods=20,freq='24H')
print(d1)
```

```python
import pandas as pd
d1=pd.date_range(start='01-01-2021',end='01-02-2021',freq='5s')
print(d1)
```

```python
date=pd.Series(pd.date_range(start='01-01-2021',end='01-02-2021',freq='5s'))
print(date)
```

```python
date=pd.Series(pd.date_range(start='01-01-2021',end='01-02-2021',freq='24H'))
print(date)
```
