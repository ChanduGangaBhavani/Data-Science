```python
df['Thal'].unique()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Thal']=label.fit_transform(df['Thal'])
df['ChestPain']=label.fit_transform(df['ChestPain'])
df['AHD']=label.fit_transform(df['AHD'])
```

```python
df.head()
```

```python
x=df.drop("AHD",axis=1)
```

```python
x
```

```python
y=df['AHD']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x,y)
```

```python
df.head()
```

```python
model.predict([[63,1,3,145,233,1,2,150,0,2.3,3,0.0,0]])
```

```python
y_predict=model.predict(x_test)
```

```python
df.info()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_predict)*100
```

### Apply logistic Regression to diabetes data and find accuracy and confusion matrix

```python
data=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Classification/diabetes.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x=data[data.columns[:-1]]
```

```python
x
```

```python
y=data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x,y)
```

```python
y_predict=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_predict)*100
```

```python
confusion_matrix(y_test,y_predict)
```
