```python
import pandas as pd
```

```python
# heart disease
df=pd.read_csv("C:/Users/India/Downloads/archive (1)/heart.csv")
df.head()
```

```python
x=df.drop("target",axis=1)
x.head()
```

```python
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
help(RandomForestClassifier())
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=10)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
accuracy_score(y_test,y_pred)
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV

#no of trees in random forest
n_estimators=[int(x) for x in np.linspace(start=200,stop=2000,num=10)]
#max_features in no of features at every split
max_features=['auto','sqrt']
#max_depth no of levels of the tree
max_depth=[int(x) for x in np.linspace(100,500,num=11)]
random_grid={"n_estimators":n_estimators,"max_features":max_features,"max_depth":max_depth}
rfc=RandomizedSearchCV(estimator=model,param_distributions=random_grid,n_iter=100,cv=3,random_state=42,verbose=2,n_jobs=-1)
rfc.fit(x_train,y_train)
print(rfc.best_params_)
```

```python
help(RandomizedSearchCV)
```

```python
model=RandomForestClassifier(n_estimators=600,max_features='auto',max_depth=380)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cv_score=cross_val_score(model,x,y,cv=10)
print(cv_score)
```

```python
## Random forest regressor
## It predicts the values by taking the average of all output of all trees
data=pd.read_csv("https://raw.githubusercontent.com/AP-State-Skill-Development-Corporation/Datasets/master/Regression/Salary_Data.csv")
```

```python
data.head()
```

```python
x=data['YearsExperience'].values.reshape(-1,1)
```

```python
y=data['Salary'].values.reshape(-1,1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
model.score(y,y_pred)*100
```
